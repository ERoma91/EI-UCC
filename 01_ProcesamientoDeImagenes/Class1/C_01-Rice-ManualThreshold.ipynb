```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

# Functions

```python
def f_howIsImg(img):
    print('Shape = ', img.shape)
    print('MaxValue = ', np.max(img))
    print('MinValue = ', np.min(img))

def f_thresholdSegmet(ch, t):
    (X, Y) = ch.shape
    imgOut = np.zeros(ch.shape)
    area = 0
    for i in range(X):
        for j in range(Y):
            if ch[i, j] > t:
                imgOut[i, j] = 255
                area =+ 1
    print('Area = ', area)
    return imgOut
```

## 1. Image Read

```python
imgSrcName = 'onerice.bmp'
imgSrc = cv2.imread(imgSrcName)

plt.figure()
plt.imshow(imgSrc)

f_howIsImg(imgSrc)
```

## First Chanel of Image

```python
imgCh1 = imgSrc[:, :, 0]
f_howIsImg(imgCh1)
plt.imshow(imgCh1, cmap='gray')
plt.show()
```

## Threshold Segmentation

```python
imgSegmentated = f_thresholdSegmet(imgCh1, 120)
f_howIsImg(imgSegmentated)
plt.imshow(imgSegmentated)
plt.show()
```

# Multiple Segmentation

```python
imgSrc = cv2.imread('rices.png')
f_howIsImg(imgSrc)
plt.imshow(imgSrc)
plt.show()
```

```python
imgCh1 = imgSrc[:,:,0]
f_howIsImg(imgCh1)
plt.imshow(imgCh1, cmap='gray')
plt.show()
```

# Homogen background
We'll substract the minimum value of the row to each pixel

```python
(x,y) = imgCh1.shape
imgOut = np.zeros(imgCh1.shape)

for i in range(x):
    xMin = np.min(imgCh1[i,:])
    #print(xMin)
    imgOut[i,:] = imgCh1[i,:] - xMin

f_howIsImg(imgOut)
plt.imshow(imgOut, cmap='gray')
plt.show()

```

```python
imgSegmentated = f_thresholdSegmet(imgOut, 60)
plt.imshow(imgSegmentated)
plt.show()
```
